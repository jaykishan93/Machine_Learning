```python
import matplotlib.pyplot as plot
import numpy as np
import pandas as pd
import seaborn as seabornInstance
```

```python
df_data = pd.read_csv('ex1data1.txt', header=None)
plot.figure(figsize=(8,8/1.5))
seabornInstance.scatterplot(x=0, y=1, data=df_data)
```

```python
plot.figure(figsize=(8,8/1.5))
seabornInstance.distplot(df_data[0])
```

```python
a = np.array(df_data)
np.random.shuffle(a)
train, test = np.split(a, [int(a.shape[0]*1)])
```

```python
def MeanSquaredCostFunction(theta0, theta1, test):
    return(((theta0 + theta1 * test[:,0] - test[:,1])**2).sum() / (2 * test.shape[0]))
    
def gradientDesecent(theta0, theta1, train, rate):
    theta0 = theta0 - rate * (1/train.shape[0]) * (theta0 + theta1 * train[:,0] - train[:,1]).sum()
    theta1 = theta1 - rate * (1/train.shape[0]) * ((theta0 + theta1 * train[:,0] - train[:,1]) * train[:,0]).sum()
    return(theta0, theta1)
```

```python
rate = 0.01
theta0 = 0
theta1 = 0
cost = []
#print(MeanSquaredCostFunction(0, 0, train))
for i in range(1500):
    theta0, theta1 = gradientDesecent(theta0, theta1, train, rate)
    cost.append(MeanSquaredCostFunction(theta0, theta1, train))

print(theta0, theta1)
print(cost[-1])
plot.figure(figsize=(15, 10))
seabornInstance.scatterplot(x=range(len(cost)), y=cost)
```
